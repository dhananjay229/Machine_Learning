```
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```
df = pd.read_csv("/content/drive/My Drive/ML/1.01. Simple linear regression.csv")
df.head()
```

```
%matplotlib inline
plt.xlabel('GPA')
plt.ylabel('SAT')
plt.scatter(df.GPA,df.SAT,color='red',marker='*')
```

```
new_df = df.drop('GPA',axis='columns')
new_df.head()
```

```
gpa = df.GPA
gpa.head()
```

```
reg = linear_model.LinearRegression()
reg.fit(new_df,gpa)
```

```
reg.predict([[2150]])
```

```
reg.coef_
```

```
reg.intercept_
```

```
p = reg.predict(_df)
```
